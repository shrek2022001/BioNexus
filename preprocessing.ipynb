```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
def preprocess_text(text):
    # Convert text to lowercase
    text = text.lower()
    # Remove non-alphabetic characters
    text = re.sub(r'[^a-z\s]', '', text)
    # Tokenize text
    tokens = nltk.word_tokenize(text)
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]
    # Lemmatize
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    # Re-join tokens into a string
    return ' '.join(tokens)
```

```python
new_df['processed_abstract'] = new_df['abstract'].apply(preprocess_text)
```

```python
from transformers import AutoTokenizer, AutoModel
import torch

# Load BioBERT tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("dmis-lab/biobert-v1.1")
model = AutoModel.from_pretrained("dmis-lab/biobert-v1.1")

# Ensure a GPU is available for PyTorch
if torch.cuda.is_available():
    model = model.cuda()
```

```python
import torch
import torch.nn.functional as F

def generate_embeddings(text):
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=512)
    outputs = model(**inputs)
    # Extract the embeddings of the [CLS] token which is at position 0
    return outputs.last_hidden_state[:, 0, :]

# For the query, ensure it is a single tensor with the first dimension as 1
query = "cancer"
query_embedding = generate_embeddings(preprocess_text(query))

# Stack all article embeddings into a tensor
all_embeddings = torch.stack([generate_embeddings(text) for text in new_df['processed_abstract']])

def find_similar_articles(query_embedding, all_embeddings, top_k=5):
    # Compute cosine similarities
    similarities = F.cosine_similarity(query_embedding, all_embeddings, dim=1)
    # Get the indices of the top_k most similar articles
    most_similar_indices = similarities.argsort(descending=True)[:top_k]
    return most_similar_indices

similar_articles_indices = find_similar_articles(query_embedding.unsqueeze(0), all_embeddings)
```
